```python
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
import os
import pandas as pd

df = pd.DataFrame()

for x in os.listdir("./"):
    if x[-4:] == ".txt":
        with open(x) as f:
            current_file_df = pd.DataFrame(f.readlines())
            current_file_df[0] = current_file_df[0].str.replace('\n', '')
            current_file_df["year"] = x[3:-4]
            current_file_df[["name","sex","number"]] = current_file_df[0].str.split(",",3, expand=True)
            current_file_df["number"] = pd.to_numeric(current_file_df["number"])
            current_file_df["name"] = current_file_df["name"].str.lower()
            current_file_df[0]
            df = pd.concat(
                [df, current_file_df],
                ignore_index=True
            )
```

```python

# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Baby names test"),
    dcc.Graph(id='graph_ext4'),
    html.Label([
        "colorscale",
        dcc.Dropdown(
            id='colorscale-dropdown', clearable=False,
            value='plasma', options=[
                {'label': c, 'value': c}
                for c in px.colors.named_colorscales()
            ])
    ]),
    html.Label([
        "names",
        dcc.Dropdown(
            id='names-dropdown', clearable=False,
            value='name', options=[
                {'label': name, 'value': name}
                for name in sorted(list(set(df.name)))
            ])
    ]),
])
```

```python
# Define callback to update graph
@app.callback(
    Output('graph_ext4', 'figure'),
    [Input("colorscale-dropdown", "value"),
     Input("names-dropdown", "value")]
)
def update_figure(colorscale, name):
    return px.scatter(
        df[(df.name == name)], x="year", y="number", color="sex",
        color_continuous_scale=colorscale,
        render_mode="webgl", title="Baby Names"
    )
```

```python
# Run app and display result inline in the notebook
app.run_server(mode='external')
```
